```python
import pandas as pd
import numpy as np
```

```python
ac=pd.read_excel("Accidents0515.xlsx")
ac.head()
```

```python
cs=pd.read_excel("Casualties0515.xlsx")
cs.head()
```

```python
vs=pd.read_excel("Vehicles0515.xlsx")
vs.head()
```

# dropping columns that arent useful

```python
#using the accidents data
ac1=ac.drop(['Location_Easting_OSGR','1st_Road_Class','Location_Northing_OSGR','Police_Force',
             '1st_Road_Number','2nd_Road_Class','2nd_Road_Number','Pedestrian_Crossing-Human_Control',
             'Pedestrian_Crossing-Physical_Facilities','Special_Conditions_at_Site','Carriageway_Hazards',
             'LSOA_of_Accident_Location'],axis='columns',inplace=False)
ac1.head()
```

## checking if there are missing values in the accidents data

```python
print(ac1.isnull().values.any())
```

```python
#removing missing values using interpolation
ac1.interpolate(method ='linear', limit_direction ='forward') 
ac1.head()
```

```python
print(ac1.isnull().values.any())
```

# the above missing values will be removed later

```python
#dropping least important columns in the casualities data set
cs1=cs.drop(['Pedestrian_Movement','Vehicle_Reference','Casualty_Reference','Age_of_Casualty'
             ,'Bus_or_Coach_Passenger','Pedestrian_Road_Maintenance_Worker','Casualty_Type',
             'Casualty_Home_Area_Type'],axis='columns',inplace=False)
cs1.head()
```

```python
#checking if there are missing values.
print(cs1.isnull().values.any())
```

# the above casualities data frame doesnt have missing values

# dealing with the vehicles data set

```python
#the vehicles data set 
#dropping least important columns
vs1=vs.drop(['Vehicle_Reference','Towing_and_Articulation','Junction_Location','Vehicle_Leaving_Carriageway',
             'Hit_Object_off_Carriageway','Was_Vehicle_Left_Hand_Drive?','Age_Band_of_Driver','Engine_Capacity_(CC)',
             'Propulsion_Code','Driver_IMD_Decile','Driver_Home_Area_Type'],
            axis='columns',inplace=False)
vs1.head()
```

# checking for missing values

```python
print(vs1.isnull().values.any())
```

### we need to merge the three data sets to make it easier to deal with the data but first we drop the accident index in the casualities and vehicles data since it already appears in the accidents data

```python
#dropping the Accident_Index in cs dataframe
cs2=cs1.drop(['Accident_Index'],axis='columns',inplace=False)
cs2.head()
```

```python
#dropping the Accident_Index in vs dataframe
vs2=vs1.drop(['Accident_Index'],axis='columns',inplace=False)
vs2.head()
```

```python
#merging them
all_data=pd.concat([ac1,vs2,cs2],axis='columns')
all_data.head()
```

```python
all_data.columns
```

```python
all_data.isnull().values.any()
```

```python
all_data.isnull().values.any()
```

```python
#counting the number of misssing values
all_data.isnull().sum()
```

```python
all_data['Casualty_Severity'].replace(2,0,inplace=True)
all_data['Casualty_Severity'].replace(3,0,inplace=True)
all_data['Casualty_Severity'].head(20)
```
